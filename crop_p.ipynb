## Treatement of Outliers

```python
q1=df.quantile(0.25)
q1
```

```python
q3=df.quantile(0.75)
q3
```

```python
IQR = q3-q1
IQR
```

```python
n_high= q3.N+(1.5*IQR.N)
n_high
```

```python
n_low= q1.N-(1.5*IQR.N)
n_low
```

```python
import numpy as np
new_df = df.copy()
new_df['N']=np.where(new_df['N']>n_high,
                    n_high,
                    np.where(new_df['N']<n_low,
                    n_low,
                    new_df['N']
    )
)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['N'])

plt.subplot(2,2,2)
sns.boxplot(df['N'])

plt.subplot(2,2,3)
sns.distplot(new_df['N'])

plt.subplot(2,2,4)
sns.boxplot(new_df['N'])

plt.show()
```

```python
p_high= q3.P+(1.5*IQR.P)
p_high
```

```python
p_low= q1.P-(1.5*IQR.P)
p_low
```

```python
new_df['P']=np.where(new_df['P']>p_high,
                    p_high,
                    np.where(new_df['P']<p_low,
                    p_low,
                    new_df['P']
    )
)
```
